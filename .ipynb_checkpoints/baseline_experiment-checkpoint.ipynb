````python
{
  "cells": [
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "# Real-ESRGAN Baseline Qualitative Experiment\n",
        "\n",
        "This notebook runs the first experiment for the paper: **shared-prior baseline qualitative evaluation** on three domains: face, architecture, and natural scenes.\n",
        "\n",
        "It will:\n",
        "1. Check the folder structure.\n",
        "2. Run Real-ESRGAN on three domain folders.\n",
        "3. Load input/output images.\n",
        "4. Generate side-by-side comparison figures for the thesis.\n"
      ]
    },
    {
      "cell_type": "code",
      "execution_count": null,
      "metadata": {},
      "outputs": [],
      "source": [
        "import os\n",
        "import subprocess\n",
        "from pathlib import Path\n",
        "from PIL import Image\n",
        "import matplotlib.pyplot as plt\n",
        "\n",
        "project_root = Path('/root/Real-ESRGAN')\n",
        "python_exec = '/root/miniconda/envs/realesrgan310/bin/python'\n",
        "inference_script = project_root / 'inference_realesrgan.py'\n",
        "\n",
        "# Input folders\n",
        "input_root = project_root / 'paper_inputs'\n",
        "face_dir = input_root / 'face'\n",
        "arch_dir = input_root / 'architecture'\n",
        "natural_dir = input_root / 'natural'\n",
        "\n",
        "# Output folders\n",
        "output_root = project_root / 'paper_results' / 'shared_prior'\n",
        "face_out = output_root / 'face'\n",
        "arch_out = output_root / 'architecture'\n",
        "natural_out = output_root / 'natural'\n",
        "\n",
        "for d in [face_dir, arch_dir, natural_dir, face_out, arch_out, natural_out]:\n",
        "    d.mkdir(parents=True, exist_ok=True)\n",
        "\n",
        "print('Project root:', project_root)\n",
        "print('Input root:', input_root)\n",
        "print('Output root:', output_root)\n"
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "## Step 1: Put your images into these folders\n",
        "\n",
        "- `/root/Real-ESRGAN/paper_inputs/face`\n",
        "- `/root/Real-ESRGAN/paper_inputs/architecture`\n",
        "- `/root/Real-ESRGAN/paper_inputs/natural`\n",
        "\n",
        "You can put a few `.jpg` or `.png` images into each folder, then run the next cell.\n"
      ]
    },
    {
      "cell_type": "code",
      "execution_count": null,
      "metadata": {},
      "outputs": [],
      "source": [
        "def list_images(folder):\n",
        "    exts = ['*.png', '*.jpg', '*.jpeg', '*.bmp', '*.webp']\n",
        "    files = []\n",
        "    for ext in exts:\n",
        "        files.extend(folder.glob(ext))\n",
        "    return sorted(files)\n",
        "\n",
        "for name, folder in [('face', face_dir), ('architecture', arch_dir), ('natural', natural_dir)]:\n",
        "    imgs = list_images(folder)\n",
        "    print(f'{name}: {len(imgs)} image(s) found')\n",
        "    for p in imgs[:10]:\n",
        "        print('  ', p.name)\n"
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "## Step 2: Run Real-ESRGAN baseline inference\n",
        "\n",
        "This uses the original `RealESRGAN_x4plus` model as the **shared prior baseline**.\n"
      ]
    },
    {
      "cell_type": "code",
      "execution_count": null,
      "metadata": {},
      "outputs": [],
      "source": [
        "def run_realesrgan(input_dir, output_dir):\n",
        "    cmd = [\n",
        "        python_exec,\n",
        "        str(inference_script),\n",
        "        '-n', 'RealESRGAN_x4plus',\n",
        "        '-i', str(input_dir),\n",
        "        '-o', str(output_dir),\n",
        "        '--fp32'\n",
        "    ]\n",
        "    print('Running:', ' '.join(cmd))\n",
        "    result = subprocess.run(cmd, cwd=project_root, capture_output=True, text=True)\n",
        "    print(result.stdout)\n",
        "    if result.returncode != 0:\n",
        "        print(result.stderr)\n",
        "        raise RuntimeError(f'Inference failed for {input_dir}')\n",
        "\n",
        "for input_dir, output_dir in [\n",
        "    (face_dir, face_out),\n",
        "    (arch_dir, arch_out),\n",
        "    (natural_dir, natural_out),\n",
        "]:\n",
        "    if len(list_images(input_dir)) > 0:\n",
        "        run_realesrgan(input_dir, output_dir)\n",
        "    else:\n",
        "        print(f'Skip {input_dir}, no images found.')\n"
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "## Step 3: Visualize side-by-side comparisons\n",
        "\n",
        "This cell creates simple input/output panels for thesis figures.\n"
      ]
    },
    {
      "cell_type": "code",
      "execution_count": null,
      "metadata": {},
      "outputs": [],
      "source": [
        "def find_output_image(output_dir, input_name_stem):\n",
        "    candidates = list(output_dir.glob(f'{input_name_stem}*'))\n",
        "    candidates = [p for p in candidates if p.suffix.lower() in ['.png', '.jpg', '.jpeg', '.bmp', '.webp']]\n",
        "    return sorted(candidates)[0] if candidates else None\n",
        "\n",
        "def show_domain_results(domain_name, input_dir, output_dir, max_items=3):\n",
        "    inputs = list_images(input_dir)[:max_items]\n",
        "    if len(inputs) == 0:\n",
        "        print(f'No images found in {input_dir}')\n",
        "        return\n",
        "    \n",
        "    fig, axes = plt.subplots(len(inputs), 2, figsize=(10, 4 * len(inputs)))\n",
        "    if len(inputs) == 1:\n",
        "        axes = [axes]\n",
        "    \n",
        "    for row, inp in enumerate(inputs):\n",
        "        out = find_output_image(output_dir, inp.stem)\n",
        "        inp_img = Image.open(inp).convert('RGB')\n",
        "        axes[row][0].imshow(inp_img)\n",
        "        axes[row][0].set_title(f'Input: {inp.name}')\n",
        "        axes[row][0].axis('off')\n",
        "        \n",
        "        if out is not None:\n",
        "            out_img = Image.open(out).convert('RGB')\n",
        "            axes[row][1].imshow(out_img)\n",
        "            axes[row][1].set_title(f'Restored: {out.name}')\n",
        "        else:\n",
        "            axes[row][1].text(0.5, 0.5, 'Output not found', ha='center', va='center')\n",
        "        axes[row][1].axis('off')\n",
        "    \n",
        "    plt.suptitle(f'Baseline Qualitative Results: {domain_name}', fontsize=16)\n",
        "    plt.tight_layout()\n",
        "    plt.show()\n",
        "\n",
        "show_domain_results('Face', face_dir, face_out, max_items=3)\n",
        "show_domain_results('Architecture', arch_dir, arch_out, max_items=3)\n",
        "show_domain_results('Natural Scene', natural_dir, natural_out, max_items=3)\n"
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "## Step 4: Save thesis-ready figures\n",
        "\n",
        "This cell saves one figure per domain.\n"
      ]
    },
    {
      "cell_type": "code",
      "execution_count": null,
      "metadata": {},
      "outputs": [],
      "source": [
        "def save_domain_figure(domain_name, input_dir, output_dir, save_path, max_items=3):\n",
        "    inputs = list_images(input_dir)[:max_items]\n",
        "    if len(inputs) == 0:\n",
        "        print(f'No images found in {input_dir}')\n",
        "        return\n",
        "\n",
        "    fig, axes = plt.subplots(len(inputs), 2, figsize=(10, 4 * len(inputs)))\n",
        "    if len(inputs) == 1:\n",
        "        axes = [axes]\n",
        "\n",
        "    for row, inp in enumerate(inputs):\n",
        "        out = find_output_image(output_dir, inp.stem)\n",
        "        inp_img = Image.open(inp).convert('RGB')\n",
        "        axes[row][0].imshow(inp_img)\n",
        "        axes[row][0].set_title('Input')\n",
        "        axes[row][0].axis('off')\n",
        "\n",
        "        if out is not None:\n",
        "            out_img = Image.open(out).convert('RGB')\n",
        "            axes[row][1].imshow(out_img)\n",
        "            axes[row][1].set_title('Real-ESRGAN Output')\n",
        "        else:\n",
        "            axes[row][1].text(0.5, 0.5, 'Output not found', ha='center', va='center')\n",
        "        axes[row][1].axis('off')\n",
        "\n",
        "    plt.suptitle(f'{domain_name} Results under Shared Prior', fontsize=16)\n",
        "    plt.tight_layout()\n",
        "    plt.savefig(save_path, dpi=300, bbox_inches='tight')\n",
        "    plt.close(fig)\n",
        "    print('Saved:', save_path)\n",
        "\n",
        "fig_dir = project_root / 'paper_results' / 'figures'\n",
        "fig_dir.mkdir(parents=True, exist_ok=True)\n",
        "\n",
        "save_domain_figure('Face', face_dir, face_out, fig_dir / 'face_shared_prior.png', max_items=3)\n",
        "save_domain_figure('Architecture', arch_dir, arch_out, fig_dir / 'architecture_shared_prior.png', max_items=3)\n",
        "save_domain_figure('Natural Scene', natural_dir, natural_out, fig_dir / 'natural_shared_prior.png', max_items=3)\n"
      ]
    },
    {
      "cell_type": "markdown",
      "metadata": {},
      "source": [
        "## Suggested LaTeX usage\n",
        "\n",
        "You may insert the saved figure into your thesis like this:\n",
        "\n",
        "```latex\n",
        "\\begin{figure}[t]\n",
        "    \\centering\n",
        "    \\includegraphics[width=0.9\\linewidth]{figures/face_shared_prior.png}\n",
        "    \\caption{Qualitative results of the shared-prior Real-ESRGAN baseline on face images.}\n",
        "    \\label{fig:qual_face}\n",
        "\\end{figure}\n",
        "```\n"
      ]
    }
  ],
  "metadata": {
    "kernelspec": {
      "display_name": "Python 3",
      "language": "python",
      "name": "python3"
    },
    "language_info": {
      "name": "python",
      "version": "3.10"
    }
  },
  "nbformat": 4,
  "nbformat_minor": 5
}
````
